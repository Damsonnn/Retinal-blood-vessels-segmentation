```python
import cv2
import os

folders = ["./image", './label']

for folder in folders:
    ids = os.listdir(folder)
    for image_id in ids:
        im = os.path.join(folder, image_id)
        image = cv2.imread(im)
        print(image.shape)
        size = (1024, int(1024/3504*2336))
        image = cv2.resize(image, size)
        print(image.shape)
        cv2.imwrite(im, image)
```
